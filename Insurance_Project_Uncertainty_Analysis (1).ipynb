# Risk classification using the claim incurred and premium

```python
df['Avrg_claim'] = df['clm.incurred']/df['clm.count']
df['percentage_of_premium'] = df['clm.incurred']/df['premium']*100
```

```python
 df['percentage_of_premium'] = df['percentage_of_premium'].fillna(0)
 df['Avrg_claim'] = df['Avrg_claim'].fillna(0)
```

```python
df.loc[df['percentage_of_premium'] <= 25, 'risk'] = 'Low_risk'
df.loc[(df['percentage_of_premium'] > 25) & (df['percentage_of_premium'] <= 50), 'risk'] = 'medium_risk'
df.loc[(df['percentage_of_premium'] > 50) & (df['percentage_of_premium'] <= 75), 'risk'] = 'high_risk'
df.loc[(df['percentage_of_premium'] > 75) & (df['percentage_of_premium'] <= 100), 'risk'] = 'very_high_risk'
df.loc[df['percentage_of_premium'] > 100, 'risk'] = 'extremely_high_risk'

print(df)
```

# Resampling of minority samples

```python
from sklearn.utils import resample
df_lowrisk = df[df.risk == 'Low_risk']
df_highrisk = df[df.risk != 'Low_risk']
df_minority_upsampled = resample(df_highrisk,replace=True,n_samples=15000)
new_df = pd.concat([df_lowrisk, df_minority_upsampled])
from sklearn.utils import shuffle
new_df = shuffle(new_df)
```

```python
new_df.info()
```

# Dummy variable creation

```python
new_df=pd.get_dummies(data=new_df, columns=['nb.rb','driver.gender', 'marital.status','body.code','fuel.type'],drop_first=True)
```

```python
def risk(x):
    if x == 'Low_risk':
        return 0
    if x == 'medium_risk':
        return 1
    if x == 'high_risk':
        return 2
    if x == 'very_high_risk':
        return 3
    else:
        return 4
new_df["risk"] = new_df["risk"].apply(risk)
```

# Importing the ML Algorithm for Analysis

```python
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from xgboost.sklearn import XGBClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier
from sklearn import metrics
```

# Cross validation of algorithms

```python
name = list(new_df.columns)
models = []
models.append(('LG', LogisticRegression()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('RDF', RandomForestClassifier()))
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10)
    cv_results = model_selection.cross_val_score(model, X, y, cv=kfold,scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

# Algorithm comparison using accuracy

```python
fig = plt.figure()
fig.suptitle('Algorithm Accuracy Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

# splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

# Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import brier_score_loss
from sklearn.preprocessing import label_binarize
from sklearn.metrics import make_scorer
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Define the model
log_reg = LogisticRegression(random_state=0)

# Define the parameters for the grid search
param_grid_lr = {
    'C': [0.01, 0.1, 1],
    'penalty': ['l2'],
    'solver': ['lbfgs'],
}

# Perform the grid search
grid_search_lr = GridSearchCV(estimator=log_reg, param_grid=param_grid_lr, cv=5)
grid_search_lr.fit(X_train, y_train.ravel())  # ravel() is used to convert y_train to a 1D array

# Print the best parameters and the best score
print(f"Best Parameters: {grid_search_lr.best_params_}")
print(f"Best Score: {grid_search_lr.best_score_}")
```

```python
import numpy as np
from sklearn.metrics import brier_score_loss
from sklearn.preprocessing import label_binarize
from sklearn.metrics import make_scorer

# Binarize the output for calculating Brier score later
classes = np.unique(y_train)
y_train_bin = label_binarize(y_train, classes=classes)
y_test_bin = label_binarize(y_test, classes=classes)

# Perform the grid search
grid_search_lr = GridSearchCV(estimator=log_reg, param_grid=param_grid_lr, cv=5)
grid_search_lr.fit(X_train, y_train) # Fit model using the original y_train

# Predict the probabilities on the test set
y_prob = grid_search_lr.predict_proba(X_test)

# Calculate the Brier score for each class
brier_scores = [brier_score_loss(y_test_bin[:, i], y_prob[:, i]) for i in range(len(classes))]

# Calculate the average Brier score
avg_brier_score = np.mean(brier_scores)

print(f"Average Brier Score: {avg_brier_score}")
```

```python
y_lr = grid_search_lr.predict(X_test)
print(metrics.classification_report(y_test,y_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define the model
rf = RandomForestClassifier(random_state=0)

# Define the parameters for the grid search
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Perform the grid search
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5)
grid_search_rf.fit(X_train, y_train)

# Print the best parameters and the best score
print(f"Best Parameters: {grid_search_rf.best_params_}")
print(f"Best Score: {grid_search_rf.best_score_}")
```
